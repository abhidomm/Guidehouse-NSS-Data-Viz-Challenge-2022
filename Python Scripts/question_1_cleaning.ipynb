```python
import os
import math
import pandas as pd
import string
import numpy as np
nss_data_path = 'C:/Users/adommalapati002/OneDrive - Guidehouse/Documents/NSS Data'
os.chdir('C:/Users/adommalapati002/OneDrive - Guidehouse/Documents/NSS Data')
os.getcwd()
```

```python
all_years = list()
for dir in os.listdir()[5:]:
    for filename in os.scandir(nss_data_path + '/' + dir):
        if filename.is_file():
            if '18d' in filename.name or 'Table18' in filename.name or 'Table 18' in filename.name:
                df_15 = pd.read_excel(filename, skiprows=4)
                if df_15.columns[0] != "Characteristic":
                    df_15 = pd.read_excel(filename, skiprows=5)
                #print(filename.name)
                #print(df_15.columns)
                df_15 = df_15.rename(columns={"Unnamed: 2": "Principal Applicants"})
                df_15 = df_15.replace("-", 0)
                df_15 = df_15.replace("D", 0)
                df_15 = df_15.replace(np.nan, 0)
                df_15 = df_15.drop_duplicates(subset=['Characteristic'], keep='first') 
                all_Characteristic_vals = ["Total", "Female", "Male", "Under 16 years", "Under 16", "16 to 20 years", "Age 16 to 20", "Age 21 and over", "21 years and over", "Married", "Married 1", "Married3", "Single", "Single 2", "Widowed", "Divorced/separated", "Divorced/Separated"]
                all_field_vals = ["Total", "Principal Applicants1", "Principal Applicants", "Spouses", "Spouses2", "Children"]
                values_dict = {}
                values_dict["Year"] = dir
                for i in df_15["Characteristic"]:
                    if str(i).strip() in all_Characteristic_vals:
                        for j in df_15.columns.difference(['Characteristic', 'Siblings', 'Other']):
                            if j in all_field_vals:
                                val = df_15[df_15["Characteristic"] == i][j]
                                if str(i).strip() == "Divorced/separated":
                                    values_dict["Divorced/Separated" + " " + j.rstrip(string.digits)] = df_15[df_15["Characteristic"] == i][j]
                                elif str(i).strip() in ["Under 16 years", "Under 16"]:
                                    values_dict["Under 16" + " " + j.rstrip(string.digits)] = df_15[df_15["Characteristic"] == i][j]
                                elif str(i).strip() in ["Age 16 to 20", "16 to 20 years"]:
                                    values_dict["16 to 20" + " " + j.rstrip(string.digits)] = df_15[df_15["Characteristic"] == i][j]
                                elif str(i).strip() in ["Age 21 and over", "21 years and over"]:
                                    values_dict["21 and over" + " " + j.rstrip(string.digits)] = df_15[df_15["Characteristic"] == i][j]
                                else:
                                    i_name = i.rstrip(string.digits)
                                    values_dict[i_name.strip() + " " + j.rstrip(string.digits)] = df_15[df_15["Characteristic"] == i][j]
                all_years.append(values_dict)

                
```

```python
print(all_years[15])
```

```python
all_df = pd.DataFrame(all_years)
```

```python
def function(x):
    if isinstance(x, str):
        return(float(x))
    elif x.empty:
        return np.nan
    elif math.isnan(x):
        return np.nan
    else:
        return float(x)

# Applying function to all the cells
result = all_df.apply(np.vectorize(function))
```

```python
result = result.fillna(0)
final_df = result.astype(int) 
```
